```python
import cmath
import numpy as np
import matplotlib.pyplot as plt
```

We want to show following identity:
$$
\mathrm{atanh}(\sqrt{x + 1}) -  \mathrm{lim_{x \rightarrow \inf} \, atanh}(x) = \mathrm{asinh}(\sqrt{\frac{1}{x}})
$$

It follows that 
$$
\mathrm{atanh}(\sqrt{x + 1}) - \mathrm{lim_{x \rightarrow \inf} \, atanh}(x) - \mathrm{asinh}(\sqrt{\frac{1}{x}}) = 0
$$

Let's plot this

```python
def f1(x):
    return cmath.atanh(cmath.sqrt(x + 1)) - cmath.atanh(cmath.inf)

def f2(x):
    return cmath.asinh(cmath.sqrt(1/x))


x = np.linspace(-10, 10, 1000)

f1_vals = np.array([f1(ix) for ix in x])
f2_vals = np.array([f2(ix) for ix in x])
dif = f1_vals - f2_vals
```

```python
colors = ['r', 'b']
ls = {'real': '-', 'imag': '--'}

fig, (ax1, ax2) = plt.subplots(2, sharex='all', figsize=(15, 10))

for v, l, c in zip([f1_vals, f2_vals], ['$f_1$', '$f_2$'], colors):
    ax1.plot(x, v.real, ls=ls['real'], color=c, label=f"{l} real")
    ax1.plot(x, v.imag, ls=ls['imag'], color=c, label=f"{l} imag")
    
ax2.plot(x, dif.real/np.pi, ls=ls['real'], color='k', label='real')
ax2.plot(x, dif.imag/np.pi, ls=ls['imag'], color='k', label='imag')

ax2.set_xlabel('x')
ax1.set_ylabel('$f_i(x)$')
ax2.set_ylabel('$(f_1(x) - f_2(x)) / \pi$')
ax1.legend()
ax2.legend()
plt.show()
plt.close()
```
